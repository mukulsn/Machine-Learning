```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/data ajio.csv')
a = df['product_image_url'][0]
df['aux_pid'] = 'AJIO_' + df['PID']
df = df.dropna(subset='product_image_url')
```

```python
df.shape
```

```python
import requests

img_data = requests.get(a).content
with open('image_name.jpg', 'wb') as handler:
    handler.write(img_data)
```

```python
def download_image(pid,url):
  img = requests.get(url).content
  with open(f'/content/ajio_dress_material_images/{pid}.jpg', 'wb') as handler:
    handler.write(img)
  return pid

result = [download_image(pid,url) for pid,url in zip(df['aux_pid'],df['product_image_url']) if url ]
```
